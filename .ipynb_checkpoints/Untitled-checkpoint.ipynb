```python
adf_result = adfuller(y)
print("ADF p-value:", adf_result[1])
```

```python
adf_result
```

```python
def adf_test(series):
    """h0:non stationary
       h1:stationary
            if p-value > 0.05 >>> non stationary
               stats value > critical value >>> non stationary
            if p-value<0.05 >>> stationary"""
    from statsmodels.tsa.stattools import adfuller 
    result=adfuller(series)
    print('ADF statistic: ',result[0])
    print('p_value: ',result[1])
    for key,value in result[4].items():
        print('Critical Value (%s): %.3f' %(key,value))
    if result[1]>0.05:
        print('data is non stationary')
    else:
        print('data is stationary')
```

```python
adf_test(y)
```

```python
def kpss_test(series):
    """h0:stationary
        h1:non stationary
            if p-value > 0.05 >>> stationary
            if p-value<0.05 >>> non stationary
            if stat value>critical value>>> non stationary"""
    result=kpss(series,regression='c')
    print('KPSS statistic: ',result[0])
    print('p_value: ',result[1])
    for key,value in result[3].items():
        print('Critical Value (%s):%.3f'%(key,value)) 
    if result[1]>0.05:
        print('data is stationary')
    else:
        print('data is non stationary')
```

```python
kpss_test(y)
```

```python
def ks_test_stationarity(series):
    """ho-Same distribution
    h1:Different distributions
    if pvalue> o.o5 >>> same data distribution
    if pvalue<0.05 >>> diff data distribution"""

    from scipy.stats import ks_2samp
    split=int(len(series)/2 )   #slicing indices int not want float format
    
    series_first_half=series[:split]
    series_second_half=series[split:]
    stat,p_value=ks_2samp(series_first_half,series_second_half)
    print('stat',stat)
    print('p_value: ',p_value)
    if p_value>0.05:
        print('data distribution is same ie strick stationary')
    else:
        print('data distribution is not same ie weak stationary ')
```

```python
ks_test_stationarity(y)
```

```python
plot_acf(y)
plot_pacf(y)
plt.show()
```

```python
train = y[:-12]
test = y[-12:] 
```

```python
model = SARIMAX(train,
                order=(1,1,1),
                seasonal_order=(1,1,1,52),
                enforce_stationarity=False,
                enforce_invertibility=False)

result = model.fit()

print(result.summary())
```

```python
forecast = result.forecast(steps=12)
```

```python
forecast
```

```python
rmse = np.sqrt(mean_squared_error(test, forecast))
mae = mean_absolute_error(test, forecast)

print("SARIMA RMSE:", rmse)
print("SARIMA MAE:", mae)
```

## 6.1 Create Lag Features

```python
lags = [1,2,4,8,12,26,52]

for lag in lags:
    df[f'lag_{lag}'] = df.groupby(['store_id','sku_id'])['units_sold'].shift(lag)
```

## 6.2 Rolling Mean Features

```python
df['rolling_mean_4'] = df.groupby(['store_id','sku_id'])['units_sold'].shift(1).rolling(4).mean()
df['rolling_mean_12'] = df.groupby(['store_id','sku_id'])['units_sold'].shift(1).rolling(12).mean()
```

```python
df = df.dropna()
```

## 6.4 Define Features

```python
features = [
    'store_id',
    'sku_id',
    'total_price',
    'base_price',
    'discount',
    'discount_pct',
    'is_featured_sku',
    'is_display_sku',
    'weekofyear',
    'month',
    'year',
    'quarter'
] + [f'lag_{lag}' for lag in lags] + ['rolling_mean_4','rolling_mean_12']
```

```python
target = 'units_sold'
```

## 6.5 Train-Test Split (Time Based)

```python
cutoff_date = df['week'].max() - pd.Timedelta(weeks=12)

train_df = df[df['week'] <= cutoff_date]
test_df = df[df['week'] > cutoff_date]

X_train = train_df[features]
y_train = train_df[target]

X_test = test_df[features]
y_test = test_df[target]
```

```python
model_xgb = XGBRegressor(
    n_estimators=500,
    learning_rate=0.05,
    max_depth=8,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

model_xgb.fit(X_train, y_train)
```

```python
preds = model_xgb.predict(X_test)
```

```python
rmse_xgb = np.sqrt(mean_squared_error(y_test, preds))
mae_xgb = mean_absolute_error(y_test, preds)

print("XGBoost RMSE:", rmse_xgb)
print("XGBoost MAE:", mae_xgb)
```

```python
## 6.9 WMAPE (Retail Metric)
```

```python
def wmape(y_true, y_pred):
    return np.sum(np.abs(y_true - y_pred)) / np.sum(y_true)

print("XGBoost WMAPE:", wmape(y_test, preds))


```
