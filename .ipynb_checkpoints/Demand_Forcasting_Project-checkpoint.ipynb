### Stationarity Test
#### 1)adf test  (Augmented DickeyFuller test)
#### 2)kpss test  (Kwiatkowski-Phillips-Schmidt-Shin (KPSS) test )
#### 3)ks test (Kolmogorov-Smirnov (KS) test)

```python
# Create function to check data distribution multiple time

#for ADF testing
def adf_test(series):
    """h0:Data is non stationary
       h1:Data is stationary
            if p-value > 0.05 >>> non stationary  ie. failed to reject null hypothesis
               stats value > critical value >>> non stationary
            if p-value<0.05 >>> stationary  ie. rejct null hypothesis"""
    
    from statsmodels.tsa.stattools import adfuller 
    result=adfuller(series)
    print('ADF statistic: ',result[0])
    print('p_value: ',result[1])
    for key,value in result[4].items():
        print('Critical Value (%s): %.3f' %(key,value))
    if result[1]>0.05:
        print('fail to reject null hypothesis')
        print('data is Non-Stationary')
        return 'non_statinary'
    else:
        print('reject null hypothesis')
        print('data is Stationary')
        return 'stationary'
        

#for KPSS TESTING  (Kwiatkowski-Phillips-Schmidt-Shin (KPSS) test )
def kpss_test(series):
    """h0:data is stationary
        h1:data is non stationary
            if p-value > 0.05 >>> stationary ie. failed to reject null hypothesis
            if p-value<0.05 >>> non stationary ie. reject null hypothesis
            if stat value>critical value>>> non stationary"""
    
    result=kpss(series,regression='c')
    print('KPSS statistic: ',result[0])
    print('p_value: ',result[1])
    for key,value in result[3].items():
        print('Critical Value (%s):%.3f'%(key,value)) 
    if result[1]>0.05:
        print('fail to reject null hypothesis')
        print('data is stationary')
        return 'statinary'
    else:
        print('reject null hypothesis')
        print('data is Non-Stationary')
        return 'non_statinary'

### if data is stationary then check weak or strict distribution of data 


# for ks testing (Kolmogorov-Smirnov (KS) test)
def ks_test_stationarity(series):
    """ho-Same distribution
    h1:Different distributions
    if pvalue> o.o5 >>> same data distribution  ie.fail to reject null hypothesis
    if pvalue<0.05 >>> diff data distribution ie. reject null hypothesis"""

    from scipy.stats import ks_2samp
    split=int(len(series)/2 )   #slicing indices int not want float format
    
    series_first_half=series[:split]
    series_second_half=series[split:]
    stat,p_value=ks_2samp(series_first_half,series_second_half)
    print('stat',stat)
    print('p_value: ',p_value)
    if p_value>0.05:
        print('fail to reject null hypothesis')
        print('data distribution is same ie strick stationary')
        return 'strict_stationary'
    else:
        print('reject null hypothesis')
        print('data distribution is not same ie weak stationary ')
        return 'weak_stationary'
```

```python
print('checking data is stationary or not by adf test')
adf_test(df11['units_sold'])
print('-'*40)
print('checking data is stationary or not by kpss test')
kpss_test(df11['units_sold'])
print('-'*40)
print('both test show that data is stationary so check weak or strict distribution now by using ks_test_hypothesis')
ks_test_stationarity(df11['units_sold'])

print('here data is Non stationary so we  need to make data stationary by using differencing,transformation,detrending,seasonal adjustment')
```

```python
#convert non stationary data to stationary form 

print('checking data is stationary or not by adf test after diff')
adf_test(df11['units_sold'].diff().dropna())
print('-'*40)
print('checking data is stationary or not by kpss test after diff')
kpss_test(df11['units_sold'].diff().dropna())
print('-'*40)
print('both test show that data is stationary so check weak or strict distribution now by using ks_test_hypothesis after diff')
ks_test_stationarity(df11['units_sold'].diff().dropna())
```

#### checking data has Noise or random walk

```python
import numpy as np 
import matplotlib.pyplot as plt 
from statsmodels.stats.diagnostic import acorr_ljungbox
```

#### Ljung -Box test-tocheck data has noise or random walk

```python
#create function to check data has noise or white noise
def ljungbox_test(serial):
    """Ho- there is no auto correlation    if p>0.05 ie white noise
     H1- there is correlation     ie  random walk   if pvalue<0.05 >>> reject null hypothesis

     pvalue>0.05 >>> white noise

     pvlaue<0.05 >>> random walk """
    pvalue=acorr_ljungbox(serial,lags=[52],return_df=True)['lb_pvalue'].iloc[0]
    if pvalue>0.05: 
        print(f'pvalue :- {round(pvalue,3)}')
        print('fail to reject null hypothesis')
        print('there is no auto correlation in data ie data has white noise')
        return 'white_noise'
    else:
        print(f'pvalue :- {round(pvalue,3)}')
        print('reject null hypothesis')
        print('there is auto correlation in data ie data has random walk noise')
        return 'random walk noise'


ljungbox_test(df11['units_sold'].diff().dropna())
```
